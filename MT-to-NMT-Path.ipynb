```python
import difflib
from deep_translator import GoogleTranslator

# The text I have created.
my_text = (
    "I am Raju Thapa, and I want to be an AI specialist. "
    "I aim to implement automation and develop faster-performing models. "
    "I am studying at the University of the Cumberlands, pursuing Artificial Intelligence. "
    "people are becoming more active rather than passive. "
    "aI can perform various tasks, such as autonomous driving. "
    "it can also enhance mobile phone technology, making devices better than current models."
)

# Translating to French and I am using Google Translator
translatedText = GoogleTranslator(source='en', target='fr').translate(my_text)

# Translating back to English and again using Google Translator
reTranslatedText = GoogleTranslator(source='fr', target='en').translate(translatedText)

# Comparing the original and final translation
diff = list(difflib.ndiff(my_text.split(), reTranslatedText.split()))

print("\nMy Text")
print(my_text)

print("\n French Translation")
print(translatedText)

print("\nEnglish Again")
print(reTranslatedText)

print("\Differences")
for line in diff:
    print(line)

# Writing theoutput to a file name translation difference
with open("translation_difference.txt", "w", encoding="utf-8") as file:
    file.write("Original Text:\n" + my_text + "\n\n")
    file.write("Translated to French:\n" + translatedText + "\n\n")
    file.write("Back to English:\n" + reTranslatedText + "\n\n")
    file.write("Differences:\n" + "\n".join(diff))

print("\nTranslation comparison saved to 'translation_difference.txt'.")
```
